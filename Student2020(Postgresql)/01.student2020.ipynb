```python
#1.Izdvojiti sve informacije o postojecim studenitma.
%sql SELECT * FROM DOSIJE LIMIT 10;
```

```python
#2.Izdvojiti brojeve indeksa studenata koji su polagali neki ispit
%sql SELECT * FROM ISPIT WHERE STATUS ='o' AND ocena >=5 LIMIT 10;
```

```python
#3.Izdvojiti razlicite brojeve indeksa studenata koji su polagali neki ispit.
%sql SELECT indeks FROM ISPIT WHERE STATUS ='o' AND ocena >=5 LIMIT 10;
```

```python
#4.Izdvojiti listu parova oznaka roka, ocena za sva polaganja bez ponavljanja.
%sql SELECT DISTINCT OZNAKAROKA,OCENA FROM ISPIT  ORDER BY 1 LIMIT 10; 


```

```python
#5.Izdvojiti brojeve indeksa i prezimena studenata koji se zovu ‘Nikola’ ili su rodeni u Beogradu
%sql SELECT INDEKS, PREZIME FROM DOSIJE WHERE IME = 'Nikola' or mestorodjenja = 'Beograd' LIMIT 10; 
```

```python
#6.Izdvojiti id-jeve predmeta koji nose bar 6 bodova.
%sql select id from predmet where espb >=6 limit 10

```

```python
#7.Izdvojiti sva uspesna polaganja ispita u Septembru 4 2019. godine.
%sql SELECT skgodina, oznakaroka, idpredmeta, status, ocena FROM  ispit WHERE skgodina = 2019 and oznakaroka='sep4' and status ='o' and ocena>5 limit 10;
```

```python
#8.Izdvojiti za svako uspesno polaganje indeks, ocenu i ocenu sracunatu preko bodova sa ispita, 
# kao prvi veci ceo broj od celobrojnog deljenja broja bodova sa 10.

%sql SELECT INDEKS, OCENA, (poeni-1)/10+1 "OCENA_CALC" FROM  ispit WHERE status ='o' and ocena>5 limit 10;
```

```python
#9.Izdvojiti brojeve indeksa i datum upisa studenata sa indeksom vecim od 20140000, uredene opadajuce prema datumu upisa.
#%sql SELECT pg_typeof(datupisa) FROM dosije LIMIT 10;

%sql SELECT * FROM ispit WHERE indeks >= '20140000'  order by 3 desc limit 10    ; 
```

```python
%sql SELECT * FROM dosije WHERE datupisa >= '2020-01-01'::date order by 1 desc limit 10 ;
```

```python
#10.Izdvojiti sva polaganja ispita sa ocenom 8, uredena prema datumu ispita opadajuce, a zatim prema broju indeksa rastuce
%sql SELECT * FROM ispit where ocena = 8 order by indeks asc, datpolaganja desc limit 10; 
```

```python
#11.Izdvojiti nazive predmeta koji nose izmedu 4 i 6 bodova, ukljucujuci obe granice.
%sql SELECT * FROM PREDMET WHERE  ESPB between 4 AND 6 LIMIT 10; 
```

```python
#12.Izdvojiti nazive predmeta koji ne nose izmedu 4 i 6 bodova. < and >, not (*), not between
%sql SELECT * FROM PREDMET WHERE  ESPB >4 AND ESPB <6 LIMIT 10; 


```

```python
#13.Izdvojiti identifikatore predmeta za koje postoji polaganje na kom je neki student dobio neparnu ocenu.
%sql SELECT idpredmeta, ocena  FROM ISPIT where ocena in (5,7,9)  LIMIT 10; 

```

```python
#14.Izdvojiti indekse studenata koji nisu dobili parnu ocenu na nekom ispitu.
%sql SELECT idpredmeta, ocena  FROM ISPIT where ocena not in (5,7,9)  LIMIT 10; 
```

```python
#15.Izdvojiti nazive predmeta koji u svom imenu imaju slovo ’a’.
%sql SELECT * FROM PREDMET WHERE naziv like '%a%'LIMIT 10;
```

```python
#16.Izdvojiti nazive predmeta ciji se naziv zavrsava slovom ’1’.
%sql SELECT * FROM PREDMET WHERE naziv like '%1'LIMIT 10;
```

```python
#17.Izdvojiti imena i prezimena studenata cije je ime duze od 5 karaktera, a pocinje slovom ’M’
%sql SELECT ime, prezime FROM dosije WHERE (ime like 'M____%' )LIMIT 10;
```

```python
#18.Izdvojiti razlicite nazive predmeta koji u svom nazivu imaju kao drugo slovo ’%’, a pretposlednje .
%sql SELECT * FROM predmet WHERE naziv like '_\%%\\_' LIMIT 10;
```
